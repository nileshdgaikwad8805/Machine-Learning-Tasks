Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

Importing data

```python
train = pd.read_csv('/content/fraudTrain.csv')
test = pd.read_csv('/content/fraudTest.csv')
```

dropping Unnecessary columns from train data

```python
train = train.drop(columns=['Unnamed: 0','trans_date_trans_time','cc_num','first','last','street','city','state','zip','dob','trans_num'],axis = 1)
```

```python
train.head()
```

dropping unnecessary columns for test data

```python
test = test.drop(columns=['Unnamed: 0','trans_date_trans_time','cc_num','first','last','street','city','state','zip','dob','trans_num'],axis = 1)
```

checking test data

```python
test.head()
```

creating object of LabelEncoder

```python
encoder = LabelEncoder()
```

transforming categorical data with LabelEncoder

```python
train['merchant'] = encoder.fit_transform(train['merchant'])
train['category'] = encoder.fit_transform(train['category'])
train['gender'] = encoder.fit_transform(train['gender'])
train['job'] = encoder.fit_transform(train['job'])
test['merchant'] = encoder.fit_transform(test['merchant'])
test['category'] = encoder.fit_transform(test['category'])
test['gender'] = encoder.fit_transform(test['gender'])
test['job'] = encoder.fit_transform(test['job'])
```

checking after transformation

```python
train.head()
```

dividing data into data and labels

```python
x_train = train.drop('is_fraud',axis = 1)
```

```python
y_train = train['is_fraud']
```

replacing values from 0.0 to 0

```python
y_train.replace(0.0,0)
```

```python
y_train
```

Now comes test data

```python
test.head()
```

dividing data and labels in test dataset

```python
x_test = test.drop('is_fraud',axis = 1)
```

```python
y_test = test['is_fraud']
```

creating StandardScaler object for standardization of data

```python
Scaler = StandardScaler()
```

standardizarion of data for both the tables

```python
x_train = Scaler.fit_transform(x_train)
```

```python
x_test = Scaler.fit_transform(x_test)
```

```python
y_test.shape
```

Creating LogisticRegression model

```python
LRmodel = LogisticRegression()
```

Fitting data into the model

```python
LRmodel.fit(x_train,y_train)
```

importing accuracy_score for checking accuracy

```python
from sklearn.metrics import accuracy_score
```

checking accuracy for train data

```python
LRmodel_train_accuracy = accuracy_score(y_train,LRmodel.predict(x_train))
LRmodel_train_accuracy
```

checking accuracy for test data

```python
LRmodel_test_accuracy = accuracy_score(y_test,LRmodel.predict(x_test))
LRmodel_test_accuracy
```

creating DicisionTreeClassifier model

```python
DTmodel = DecisionTreeClassifier()
```

Fitting data into the model

```python
DTmodel.fit(x_train,y_train)
```

checking accuracy score for train data

```python
DTmodel_train_accuracy = accuracy_score(y_train,DTmodel.predict(x_train))
DTmodel_train_accuracy
```

checking accuracy score for test data

```python
DTmodel_test_accuracy = accuracy_score(y_test,DTmodel.predict(x_test))
DTmodel_test_accuracy
```

creating RandomForestClassifier object

```python
RFmodel = RandomForestClassifier(max_depth = 3)
```

Fitting data into the model

```python
RFmodel.fit(x_train,y_train)
```

checking accuracy for training data

```python
RFmodel_train_accuracy = accuracy_score(y_train,RFmodel.predict(x_train))
RFmodel_train_accuracy
```

checking accuracy score for test data

```python
RFmodel_test_accuracy = accuracy_score(y_test,RFmodel.predict(x_test))
RFmodel_test_accuracy
```

checking accuracy for all the models till used

```python
print('LogisticRegression model accuracy = ',LRmodel_test_accuracy)
print('DicisionTreeClassifier model accuracy = ',DTmodel_test_accuracy)
print('RandomForestClassifier model accuracy = ',RFmodel_test_accuracy)
```
