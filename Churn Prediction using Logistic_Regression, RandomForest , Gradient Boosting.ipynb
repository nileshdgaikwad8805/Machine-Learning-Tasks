Importing Dependies

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
```

importing data

```python
data = pd.read_csv('/content/Churn_Modelling.csv')
```

checking what's into the data

```python
data.head()
```

checking statistical summary of data

```python
data.describe()
```

gaining insights like dtypes and checking nulls

```python
data.info()
```

getting columns

```python
data.columns
```

dropping unnecessary columns

```python
data = data.drop(columns = ['RowNumber','CustomerId','Surname'],axis = 1)
```

checking once again

```python
data
```

Creating Encoder object

```python
encoder = LabelEncoder()
```

using encoder to transform categorical data into numerical

```python
data['Geography'] = encoder.fit_transform(data['Geography'])
data['Gender'] = encoder.fit_transform(data['Gender'])
```

checking data after transformation

```python
data
```

segregating x and y, like data and labels

```python
x = data.drop('Exited',axis = 1)
```

```python
y = data['Exited']
```

importing StandardScaler for standardization

```python
from sklearn.preprocessing import StandardScaler
```

creating Scaler object

```python
Scaler = StandardScaler()
```

standardizing data because bringing data into same range

```python
x = Scaler.fit_transform(x)
```

checking after transformation

```python
x
```

importing model for diving data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

checking shape after splitting

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

creating Logistic Regression model

```python
LRmodel = LogisticRegression()
```

fitting data into model

```python
LRmodel.fit(x_train,y_train)
```

importing accuracy score te chech accuracy

```python
from sklearn.metrics import accuracy_score
```

checking accuracy for train data

```python
LRmodel_train_accuracy = accuracy_score(y_train,LRmodel.predict(x_train))
LRmodel_train_accuracy
```

checking accuracy for test data

```python
LRmodel_test_accuracy = accuracy_score(y_test,LRmodel.predict(x_test))
LRmodel_test_accuracy
```

creating Random forest Classifier object

```python
RFmodel = RandomForestClassifier(max_depth = 3)
```

fitting data into the model


```python
RFmodel.fit(x_train,y_train)
```

checking accuracy for training data

```python
RFmodel_train_accuracy = accuracy_score(y_train,RFmodel.predict(x_train))
RFmodel_train_accuracy
```

checking accuracy for testing data

```python
RFmodel_Test_accuracy = accuracy_score(y_test,RFmodel.predict(x_test))
RFmodel_Test_accuracy
```

creating gradient Boosting Classifier object

```python
GBCmodel = GradientBoostingClassifier(n_estimators = 100)
```

fitting data into the model

```python
GBCmodel.fit(x_train,y_train)
```

checking accuracy for training data

```python
GBCmodel_train_accuracy = accuracy_score(y_train,GBCmodel.predict(x_train))
GBCmodel_train_accuracy
```

checking data for testing data

```python
GBCmodel_test_accuracy = accuracy_score(y_test,GBCmodel.predict(x_test))
GBCmodel_test_accuracy
```

printing all the accuracies of all the models

```python
print('LogisticRegression model accuracy = ',LRmodel_test_accuracy)
print('RandomForest model accuracy = ',RFmodel_Test_accuracy)
print('GradientBoosting model accuracy = ',GBCmodel_test_accuracy)
```
